## Part 3:  异常值处理

### 异常值处理
异常值是一些少数的值完全偏离绝大多数的值，或者不符合常识的值。如下图所示：

```python
from IPython.display import Image
Image(filename='outliers.jpg')
```

判读一个值是否是异常值，我们经常使用Tukey IQR方法。[IQR](http://sphweb.bumc.bu.edu/otlt/mph-modules/bs/bs704_summarizingdata/bs704_summarizingdata7.html)

```python
Image(filename='tukeyiqr.jpg')
```

**根据上面我们对数据的分析，我们可以看到，有可能出现异常值的分别是age、fnlwgt以及hours_per_week**

```python
# 定义函数，找出异常值的索引及其值。
def find_outliers_tukey(x):
    q1 = np.percentile(x, 25)
    q3 = np.percentile(x, 75)
    iqr = q3-q1 
    floor = q1 - 1.5*iqr
    ceiling = q3 + 1.5*iqr
    outlier_indices = list(x.index[(x < floor)|(x > ceiling)])
    outlier_values = list(x[outlier_indices])

    return outlier_indices, outlier_values
```

```python
ag_outlier_indices, ag_outlier_values = find_outliers_tukey(df['age'])
print (ag_outlier_values)
```

```python
hr_outlier_indices, hr_outlier_values = find_outliers_tukey(df['hours_per_week'])
print (hr_outlier_values)
# 从下面的结果可以看出每周的工作时间分布很零散，所以如果对每周工作时间进行异常值处理会让数据丢失很多
#有用信息，所以在此不对每周工作时间做异常值处理
```

```python
fl_outlier_indices, fl_outlier_values = find_outliers_tukey(df['fnlwgt'])
print (fl_outlier_values)
```

## Part4 数据转换

### 独热编码

需要进行独热编码的特征有'workclass', 'marital_status', 'occupation', 'relationship', 'race', 'sex', 'native_country'。

```python
%matplotlib inline
```

```python
df['fnlwgt'].plot()
```

```python
df['capital_gain'].plot()
```

```python
df['capital_loss'].plot()
```

```python
# 创建独热编码的特征列表
todummy_list = ['workclass', 'marital_status', 'occupation', 'relationship', 'race', 'sex', 'native_country']
```

```python
# 独热编码转换函数
def dummy_df(df, todummy_list):
    for x in todummy_list:
        dummies = pd.get_dummies(df[x], prefix=x, dummy_na=False)
        df = df.drop(x, 1)
        df = pd.concat([df, dummies], axis=1)
    return df
```

```python
df = dummy_df(df, todummy_list)
```

```python
df.head() # 独热编码后的结果
```

```python
df.shape[1]
```

### 大数对数处理  
对于特征值不在同一个数量级的特征值，我们需要对其进行对数转换。

```python
# 将capital_gain 和 capital_loss 的值取对数。
skewed = ['capital_gain', 'capital_loss']
df[skewed] = df[skewed].apply(lambda x: np.log(x + 1))
```

```python
df.head(50)
```

### 特征值归一化  
为了体现对每一个特征值的公平，我们需要将所有的特征缩放到同样的范围，即我们需要对其进行特征值缩放。

```python
# 导入sklearn.preprocessing.StandardScaler
from sklearn.preprocessing import MinMaxScaler

# 初始化一个 scaler，并将它施加到特征上
scaler = MinMaxScaler()
numerical = ['age', 'fnlwgt', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week']
df[numerical] = scaler.fit_transform(df[numerical])
```

## Part 5:  模型表现比较

### 用处理后的数据建模

```python
# 构建逻辑回归模型
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score

def find_model_perf(X_train, y_train, X_test, y_test):
    model = LogisticRegression()
    model.fit(X_train, y_train)
    y_hat = [x[1] for x in model.predict_proba(X_test)]
    auc = roc_auc_score(y_test, y_hat)
    
    return auc
```
